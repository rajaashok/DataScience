### Introduction to Logistic Regression

    Logistic Regression is used to solve classification problems
        Binary Classification -- If number of output classes is two
        Multi Label Classification or Multi Class Classification -- If number of output classes is more than two
    
    Examples include: Spam/Ham email classification, Website Classification, Customer churn Yes/No, Cancer diagnosis Yes/No. etc.
    
#### Although the term Regression sounds like it is predicting a continous variable, but the Logistic Regression is used to solve classification problems

###  Model Evaluation

```python
# we can use confusion matrix to evaluate our model
Image(filename='confusion_matrix.png')
```

```python
# number of success cases:

# Accuracy = (TP+TN)/total
(100 + 50)/165
```

```python
# number of failure cases:
# Error rate or misclassification rate

# Error rate = (FP + FN)/total
(10 + 5)/165
```

```python
# Precision (P) is defined as the number of true positives (T_p) over the
# number of true positives plus the number of false  positives (F_p).

100/(100+10)
```

```python
# Recall (R) is defined as the number of true positives (T_p)
# over the number of true positives plus the number of false negatives (F_n).

100/(100+5)
```
