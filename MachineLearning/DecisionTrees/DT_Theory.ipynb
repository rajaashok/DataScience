## Decision Trees

    Easy to understand and easy to explain model in Machine Learning
    Predictor variables are used build a tree(model), which eventually used to predict the target variable

### Terminology

    -> Root node - Starting point of the decision tree
    -> Edge - Each branch is referred as edge
    -> Leaf node - Final decision making nodes
    -> Node - Each node is basing a condition
    

### Under the hood, the mathematical terms involved are Entropy and Information Gain. These are used to best split the data

#### Advantages

    Easy to interpret and explain 
    Works with missing data
    Sensitive to local variations
    Fast
    
#### Disadvantages 

    Not good with large predictors
    Limited accuracy and bias influences
    
#### Ideal use cases

    Credit card approvals
    Legal decisions
    Loan application
