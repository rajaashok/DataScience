##  Unsupervised Learning

    Used to explore hidden insights in the data
    No labels
    Mainly used for clustering and dimensionality reduction
    Clustering - popular method to group data into subsets based on similarity
    Main clustering algorithms:
        1. K-means
        2. Hierarchical clustering (Agglomerative)
        3. DBSCAN  (Density based Spatial Clustering of Applications with Noise)
    Example of clustering include:
        grouping of documents, music and movies
        finding customers that share similar interests based on common purchase behaviors

##  K-means

    Easy to implement, and computationally very efficient compared to other clustering algorithms
    Works on the either measure of centroid(average of similar points) or medoid(most representative or most frequently occuring point) 
    

```python
from IPython.display import Image
Image(filename='k-means.png')
```

####  When we are applying K-means using distance metric, make sure all the data points are scaled using StandardScalar, Min-Max scalar, etc

### Disadvantages

    Find the best K  
        can be overcomed by elbow method
        
    Initial centroid changes the shape of clusters 
        k-means++ algorithm to find the best initial centroid overcomes it
        
    K-means output is always spherical 
        DBSCAN clustering method can be useful here
    
    One or more clusters can be empty
        Can be overcomed by K-medoids or fuzzy C-means algorithm
