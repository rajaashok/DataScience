## Pros
            
    1. Very simple
    2. Works with any number of classes
    3. Easy to add more data
    4. Only 'K' and distance metric are the parmaters
    
## Cons
            
    1. High prediction cost - Lot of time to compute distance if the data is huge
    2. Not good for high dimensional data
    3. Categorical features won't work well
