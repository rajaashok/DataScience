## Support Vector Machines

The goal of a support vector machine is to find  the optimal separating hyperplane which maximizes the margin of the training data.

#### Key points

    One of the supervised learning algorithm
    Used for both classification and regression problems
    Can able to model highly complex relationships
    Works well with lot of features
    
    Target: Seperate the points of the multiple categories with a hyper plane, with as wide gap as possible
