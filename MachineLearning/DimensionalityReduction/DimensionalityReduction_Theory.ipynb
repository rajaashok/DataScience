    Feature Extraction
    Feature Selection

### Main algorithms used for Feature Selection

    Variance Threshold
    SelectKBest
    SelectPercentile
    
    SelectKBest and SelectPercentile uses a scoring function:
    
        For Regression - f_regression, mutual_info_regression
        For Classification - chi2, f_classif, mutual_info_claasif
        
    Recursive Feature Elimination
    SelectFromModel

### Main algorithms used for Feature Extraction

    PCA - Pricipal component analysis
    LDA - Linear Discriminant Analysis
    Kernel PCA 
    t_SNE
